**Converting Colour Image to Grayscale**

```python
def color2gray(img_rgb):

    r, g, b = img_rgb[:,:,0], img_rgb[:,:,1], img_rgb[:,:,2]
    gray_img = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray_img
```

```python
gray1 = color2gray(img1_rgb)
# gray2 = color2gray(img1_rgb)
# gray3 = color2gray(img1_rgb)
# gray4 = color2gray(img1_rgb)
```

```python
plt.imshow(gray1, cmap = "gray")
plt.savefig("Flag_in_gray_scale.png")
```

```python
max_val = np.max(np.round(gray1))
min_val = np.min(np.round(gray1))
print("The range of gray scale values for the image lies between %d and %d"%(min_val,max_val))
```
